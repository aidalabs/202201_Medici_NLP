#### 1.3.1 Kkma
- 서울대학교 지능형 데이터 시스템(IDS) 연구소에서 개발함

- 네이버 영화 리뷰 NSMC 데이터를 이용

```python
from Korpora import Korpora

nsmc = Korpora.load("nsmc", force_download=True, root_dir=DATA_PATH)
```

- NSMC 데이터 전처리
  - NSMC에 포함된 영화 리뷰들을 순수 텍스트 형태로 지정된 디렉토리에 저장한 후 사용

```python
import os

def write_lines(path, lines):
  with open(path, 'w', encoding='utf-8') as f:
    for line in lines:
      f.write(f'{line}\n')

write_lines("{}{}".format(DATA_PATH, "nsmc_train.txt"), nsmc.train.get_all_texts())
write_lines("{}{}".format(DATA_PATH, "nsmc_test.txt"), nsmc.test.get_all_texts())
```

- BPE 토크나이저 구축
  - 문자 단위가 아니라 유니코드 바이트 단위로 어휘집합을 구축하고 토큰화 수행
  - 전 세계 대부분의 글자는 유니코드로 표현할 수 있으므로 유니코드 바이트를 기준으로 BPE를 사용하면 미등록 토큰 문제에서 비교적 자유로울 수 있음

```python
import os
os.makedirs("{}{}".format(DATA_PATH, "bbpe"))   # 바이트 단위 BPE 어휘 집합 구축결과를 저장할 디렉토리 생성
```

```python
# 어휘 집합 구축
from tokenizers import ByteLevelBPETokenizer

bytebpe_tokenizer = ByteLevelBPETokenizer()
bytebpe_tokenizer.train(
    files = ["{}{}".format(DATA_PATH, "nsmc_train.txt"), "{}{}".format(DATA_PATH, "nsmc_test.txt")],    # 학습 말뭉치를 리스트 형태로 넣기
    vocab_size = 10000,           # 어휘 집합의 크기 조절
    special_tokens = ["[PAD]"]     # 특수 토큰 추가
)
bytebpe_tokenizer.save_model("{}{}".format(DATA_PATH, "bbpe"))
```

- 생성된 파일의 내용
  - vocab.json : BPE 어휘 집합
  - merges.txt : 바이그램 쌍의 병합 우선 순위

### 2.2 토큰화
- BPE 기반 입력값 만들기
  - 앞 단계에서 만든 vocab.json, merges.txt 파일이 있어야 함

```python
from transformers import GPT2Tokenizer

tokenizer_gpt = GPT2Tokenizer.from_pretrained("{}{}".format(DATA_PATH, "bbpe"))
tokenizer_gpt.pad_token = "[PAD]"
```

```python
# GPT 토크나이저 토큰화 결과 맛보기
sentences = [
             "아 더빙.. 진짜 짜증나네요 목소리",
             "흠...포스터보고 초딩영화줄....오버연기조차 가볍지 않구나",
             "별루 였다..",
]
tokenized_sentences = [tokenizer_gpt.tokenize(sentence) for sentence in sentences]
tokenized_sentences
```

```python
batch_inputs = tokenizer_gpt(
    sentences,
    padding="max_length",       # 문장의 최대 길이에 맞춰 패딩
    max_length = 12,            # 문장의 토큰 기준 최대 길이
    truncation = True,          # 문장 잘림 허용 옵션
)
```

- 위의 코드를 실행하면 input_ids, attention_mask가 생성됨
  - input_ids: 토큰화 결과를 이용하여 각 토큰을 인덱스로 바꾼 것. 어휘 집합(vocab.json)을 보면 각 어휘 순서대로 나열되어 있는데 이 순서가 인덱스가 됨
  - attention_mask: 일반 토큰이 자리한 곳(1)과 패딩 토큰이 자리한 곳(0)을 구분하여 알려주는 장치
